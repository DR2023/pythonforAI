# Creating Arrays:

###1)Write a NumPy program to create a 3x3 identity matrix with diagonal elements [2,4,6] and stack it vertically and horizontally.
#### Sample output:
image.png

```python
import numpy as np
#A = np.eye(3)
B = [2,4,6]
A = np.diag(B) # adds diagonal elements and creates identity matrix
print(A)
print(np.vstack((A,A))) # stacked vertically
print(np.hstack((A,A))) # stacked horizontally
```

### 5)Write a NumPy program to create a 4x4 array with random integer values. Create an array by swapping first and last rows from the matrix.
image.png

```python
myarr = np.random.randint(0,100,size=(4,4),dtype=int)
print(myarr) # Created a 4x4 array
C = myarr[0,:]
D = myarr[-1,:]
myarr1 = np.copy(myarr)
myarr1[-1,:]=C
myarr1[0,:]=D
print(myarr1) # Created Swapped horizontal rows and columns
```
